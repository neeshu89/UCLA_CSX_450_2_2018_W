## Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(customer_final_df)
```

### Scree Plot

Look for "the bend". This will tell you how many components to keep.

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
customer_final_df.cov()
```

```python
import lib.viz_helper as viz
pca_df = viz.pca_results(customer_final_df, pca)
```

We're finding the hidden (latent) variables in the data
- The principal component is it's GroceryStore-ness
- The second component is its Restaurant-ness
- The next component tells you if a places is more veggie or deli heavy
- The next component tells you if its more frozen or fresh heavy
- The next component tells us if it's more detergents/paper vs milk products
- The last tells us if its more grocery vs milk/detergent

```python
pca = PCA(2)
pca.fit(customer_final_df)
```

```python
customer_pca_df = pd.DataFrame(pca.transform(customer_final_df), 
                               columns=['Dim 1', 'Dim 2'],
                               index=customer_final_df.index)
sample_pca_df = pd.DataFrame(pca.transform(sample_df), 
                             columns=['Dim 1', 'Dim 2'],
                             index=sample_df.index)
```

```python
fig = plt.figure(figsize=(12,6))
fig.add_subplot(121)
plt.title("Original Data")
sns.heatmap(sample_df, annot=True, cbar=False, square=True)
fig.add_subplot(122)
plt.title("PCA transformed Data")
sns.heatmap(sample_pca_df, annot=True, cbar=False, square=True)
```

```python
sample_df.plot(kind='bar', figsize=(12,8))
plt.title("Original Data")
_ = plt.xticks(range(5),['Sample 1','Sample 2','Sample 3', 'Sample 4','Sample 5'])

sample_pca_df.plot(kind='bar', figsize=(12,8))
plt.title("PCA transformed Data")
_ = plt.xticks(range(5),['Sample 1','Sample 2','Sample 3', 'Sample 4','Sample 5'])
```
